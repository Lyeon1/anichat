## 전체 문장 길이 구하기

```python
import wave
import contextlib
```

```python
acc = 0
for i in tqdm(total):

 with contextlib.closing(wave.open(i,'r')) as f:
  frames = f.getnframes()
  rate = f.getframerate()
  duration = frames / float(rate)
  #print(duration)
  acc += duration #누적값

#결과 출력
print("전체 문장 길이 = 약 {:.1f}분({}초)".format(acc/60, acc))
print("파일 당 평균 문장 길이 = 약 {:.3f}초".format(acc/len(total)))
```
